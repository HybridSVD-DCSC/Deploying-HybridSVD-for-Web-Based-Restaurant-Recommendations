CategorySearch must be ran first, this will create the restaurant_gmap_ids json that is needed to filter out restaurants. Then you must run n_core-rest.ipynb, this will create the 4 matrices for training (explained in that notebook). Afterwards a item-item matrix (also known as Z-matrix) can be generated.

Each creates the data needed in the next notebook so they must be ran in that order.

Furthermore, all folders and data can be deleted except anything that sits in the "Data" folder because this is the raw data from which the matrices are constructed.
